```python
import os 
import numpy as np
import pandas as pd

ONTO_PATH = "/data2/zhoujb/project/cowpea_project/rawData/geneOntology/"
GOKEGG_PATH = "/data2/zhoujb/project/cowpea_project/basedXPXLR/ML/GOKEGG"
```

```python
def getsswDf(file_path, trait_name):
    ssw_list = []
    with open(file_path) as f:
        for line in f:
            line = line.rstrip()
            line = line[4:]
            if line.startswith("Chr"):
                ssw_list.append(line)

    all_gene_table = pd.read_table(os.path.join(ONTO_PATH, "feature_table_gene_id.txt"))

    ssw_df = pd.DataFrame(columns=all_gene_table.columns)
    for item in ssw_list:
        ssw_item = item.split("_")
        ssw_start = int(ssw_item[1])
        ssw_end = int(ssw_item[2])

        tmp_data = all_gene_table[all_gene_table['Chr']==ssw_item[0]].copy()
        ssw_tmp_df = pd.DataFrame(columns=all_gene_table.columns)
        for index in tmp_data.index:
            start = int(tmp_data.loc[index, "start"])
            end = int(tmp_data.loc[index, "end"])
            if (ssw_start < start) and (ssw_end > end):
                ssw_tmp_df.loc[len(ssw_tmp_df)] = tmp_data.loc[index]

        ssw_tmp_df["ssw_item"] = item
        ssw_df = pd.concat([ssw_df, ssw_tmp_df], axis=0, ignore_index=True)

    ssw_df["traits"] = trait_name
    return ssw_df

pl_ssw_df = getsswDf("/data2/zhoujb/project/cowpea_project/basedXPXLR/ML/fs_PL/cb_rfa_cv_pl_rmse", "PL")
gnp_ssw_df = getsswDf("/data2/zhoujb/project/cowpea_project/basedXPXLR/ML/fs_GNP/cb_rfa_cv_gnp_rmse", "GNP")
tsw_ssw_df = getsswDf("/data2/zhoujb/project/cowpea_project/basedXPXLR/ML/fs_TSW/cb_rfa_cv_tsw_rmse", "TSW")

psugar_ssw_df = getsswDf("/data2/zhoujb/project/cowpea_project/basedXPXLR/ML/fs_PSugar/cb_rfa_cv_psugar_rmse", "PSugar")
pstarch_ssw_df = getsswDf("/data2/zhoujb/project/cowpea_project/basedXPXLR/ML/fs_PStarch/cb_rfa_cv_pstarch_rmse", "PStarch")
pprotein_ssw_df = getsswDf("/data2/zhoujb/project/cowpea_project/basedXPXLR/ML/fs_PProtein/cb_rfa_cv_pprotein_rmse", "PProtein")
```

```python
data_con = pd.concat([pl_ssw_df, gnp_ssw_df, tsw_ssw_df, psugar_ssw_df, pstarch_ssw_df, pprotein_ssw_df], axis=0, ignore_index=True)
data_con.to_csv(os.path.join(GOKEGG_PATH, "traits_region_gene.txt"), sep="\t", index=0)
```
