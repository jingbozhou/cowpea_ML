```python
%matplotlib inline
import os, pickle
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats

font_path = "/home/zhoujb/local/font/times.ttf"
mpl.font_manager.fontManager.addfont(font_path)
prop = mpl.font_manager.FontProperties(fname=font_path)
mpl.rcParams['font.family'] = prop.get_name()
mpl.rcParams['font.sans-serif'] = ["Times New Roman"]

VL_CROSS_PATH = "/data2/zhoujb/project/cowpea_project/basedXPXLR/Cross_Analysis_Results/VLCross/"
VC_CROSS_PATH = "/data2/zhoujb/project/cowpea_project/basedXPXLR/Cross_Analysis_Results/VCCross/"
G_CROSS_PATH = "/data2/zhoujb/project/cowpea_project/basedXPXLR/Cross_Analysis_Results/GCross/"
VCVL_CROSS_PATH = "/data2/zhoujb/project/cowpea_project/basedXPXLR/Cross_Analysis_Results/VCVLCross/"
FIG_PATH = "/data2/zhoujb/project/cowpea_project/basedXPXLR/plotFigure/"
```

```python
vl_p2_info = pd.read_table(os.path.join(VL_CROSS_PATH, "VL_TOP10_P2_info.txt"))
vl_p2_top10 = vl_p2_info.head(10)

vc_p2_info = pd.read_table(os.path.join(VC_CROSS_PATH, "VC_TOP10_P2_info.txt"))
vc_p2_top10 = vc_p2_info.head(10)

g_p2_info = pd.read_table(os.path.join(G_CROSS_PATH, "G_TOP10_P2_info.txt"))
g_p2_top10 = g_p2_info.head(10)

vcvl_p2_info = pd.read_table(os.path.join(VCVL_CROSS_PATH, "VCVL_TOP10_P2_info.txt"))
vcvl_p2_top10 = vcvl_p2_info.head(10)
```

```python
vl_p2_top10
```

```python
sns.set(font_scale=1, style="ticks", rc={'figure.figsize':(4,3)})

mpl.rcParams['font.family'] = prop.get_name()
mpl.rcParams['font.sans-serif'] = ["Times New Roman"]

ax = sns.barplot(vl_p2_top10, x="P2_donate_times", y="P2", orient="h", )
ax.bar_label(ax.containers[0])
ax.set(xlabel="Match times", ylabel="")
ax.spines[['right', 'top']].set_visible(False)

plt.savefig(os.path.join(FIG_PATH, 'VL_p2_top10.pdf'), format='pdf', bbox_inches='tight', transparent=True)
```
